```python
import os
import numpy as np
import pandas as pd
```

```python
records = []

os.chdir("../greed_search/result/")
for d in os.listdir("./"):
    if os.path.isdir(d):
        os.chdir(d)
        for f in os.listdir("./"):
            
            if f.endswith("_result.tsv"):
                df = pd.read_table(f)

                records_single = []
                for filter_params, df_filter in df.groupby(['filter_by_graph', 'filter_by_cys_pair', 'prob_threshold', 'node_num_threshold']):
                    avg_f1 = np.average(df_filter['f1'])
                    std_f1 = np.std(df_filter['f1'], ddof=1)
                    filter_by_graph, filter_by_cys_pair, prob_threshold, node_num_threshold = filter_params
                    records_single.append({
                        "filter_by_graph": filter_by_graph,
                        "filter_by_cys_pair": filter_by_cys_pair,
                        "prob_threshold": prob_threshold,
                        "node_num_threshold": node_num_threshold,
                        "f1_avg": round(avg_f1, 6),
                        "f1_std": round(std_f1, 6),
                    })
                df_single = pd.DataFrame(records_single)
                target_row = df_single.iloc[df_single['f1_avg'].argmax()]

                record = {"file": f}
                record.update(dict(target_row))
                records.append(record)
        os.chdir("..")
        
os.chdir("../../analysis/")
```

```python
pd.DataFrame(records).sort_values(by=['f1_avg'])
```
